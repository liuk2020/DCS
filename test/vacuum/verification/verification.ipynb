```python
import dcs.vmec
import matplotlib.pyplot as plt
import numpy as np
from dcs.boozerForm import BoozerForm
from dcs.vacuum import QsSurface
try:
    import matplotlib
    matplotlib.rcParams['text.usetex'] = True
except:
    pass
```

```python
file_name = "LandremanPaul2021_QA"

boozerlib = BoozerForm() 
boozerlib.read_boozmn("../datas/boozer_wout_"+file_name+".nc") 
surfBoozer = boozerlib.surface() 
iota = boozerlib.getIota()
```

```python
qsProblem = QsSurface(
    surf = surfBoozer,
    iota = iota
)
fFun, gFun, gResidual = qsProblem.funResidual()
```

```python
fig, ax = plt.subplots()
fFun.plot_plt(fig=fig, ax=ax)
```

```python
fig, ax = plt.subplots()
(gFun*(1/gFun.getRe(0,0))).plot_plt(fig=fig, ax=ax)
```

```python
fig, ax = plt.subplots()
(gResidual*(1/gFun.getRe(0,0))).plot_plt(fig=fig, ax=ax)
```
